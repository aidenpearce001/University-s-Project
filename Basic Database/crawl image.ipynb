```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
import urllib.request
```

```python
df = pd.read_csv("data/restaurant_info.csv")
```

```python
hostname = "https://www.foody.vn"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'}

dict1 = {}
for i,v in enumerate(df["Url"]):
    print(i)
    page = requests.get(hostname+v,headers=headers).text
    soup = BeautifulSoup(page)
    print(hostname+v)
    for k in soup.findAll("img", {"class": "pic-place"}):
        dict1[df['Name'][i]] = "data/images/"+k['src'].split("/")[-1]
#         print(k['src'])
        urllib.request.urlretrieve(k['src'], "data/images/"+k['src'].split("/")[-1])
```

```python
#Dict to file
with open('images_map','w+',encoding='utf-8') as f:
    for k,v in dict1.items():
        f.write(k+":"+v+'\n')
```

```python
#file to dict
dictionary = {}
with open('images_map','r',encoding='utf-8') as fd:
    for i in fd:
        dictionary[i.split(":")[0]] = i.split(":")[-1].rstrip()
```
